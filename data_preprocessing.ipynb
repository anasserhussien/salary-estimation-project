# Preprocessing steps:

    - remove any \n in the data.
    - parse and convert salary into number.
    - clean company name.
    - Seperate Location from the state.
    - create col for common skills

```python
#drop 'Unnamed: 0' column
df = df.drop('Unnamed: 0', axis= 1)

# remove \n from the dataframe
df = df.replace('\n','', regex=True)
```

## Salary preprocessing
    remove data points that miss the salary variable
    remove any string e.g (Employer Provided Salary:, Glassdoor est., Per Hour, ....)
    create col for min salary
    create col for max salary
    create col for avg salary

```python
df['Salary Estimate'].unique()
```

```python
df[df['Salary Estimate'] != '-1']
```

```python
# remove rows with -1 as salary 
df = df[df['Salary Estimate'] != '-1']

# create flags for hourly paid and estimation provided by the employer
df['hourly'] =  df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df['employer_provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0)

# remove any included strings except numbers
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
salary = salary.apply(lambda x: x.lower().replace('per hour', '').replace('employer provided salary:', '').replace('k', '').replace('$',''))

# get min and max salary based on the provided range
df['min_salary']  =salary.apply(lambda x: int(x.split('-')[0]))
df['max_salary']  =salary.apply(lambda x: int(x.split('-')[-1]))

df['avg_salary'] = (df['min_salary'] + df['max_salary']) /2

```

## Company Name preprocessing
For the Company Name we can see that it has the rating also attached to its name, but only if a rating is already provided (rating is not -1 ) so in the following step we remove the rating from the company name

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(df)
```

```python
df['company'] = df.apply(lambda row: row['Company Name'].lower()[:-3] if row['Rating'] > 0 else row['Company Name'].lower(), axis=1 )
```

## Location preprocessing
    -first we want to seperate the state from the city name
    -create another column that indicate whether the job offer in the headquarter or not

```python
# split on the comma and retrieve the two char state name
df['state'] = df['Location'].apply(lambda row: row.split(',')[1])
```

```python
df['headquarter_offer'] = df.apply(lambda row: 1 if row['Location'] == row['Headquarters'] else 0, axis=1)
```

## Examine occurances of common DS skills in Job Description
    - check whether a job description asked for specific tools such as Python, R, SQL, Excel, etc..
    - create seperate boolean column for each skill in case job desc asked for it

```python
from collections import Counter

# join all descriptions
all_job_desc = ''.join(df['Job Description'].to_list())
# lower case
all_job_desc = all_job_desc.lower()
# split on spaces to get individual words
splitted_words = all_job_desc.split()
# construct the frequency dictionary
words_count = Counter(splitted_words)

skills = ['python', 'sql', 'r', 'excel', 'ml', 'spark', 'tableau']

for skill in skills:
    print(skill + ' ->',words_count[skill] )
    
```

```python
# create new column with skill name, value is 1 if skill in desc
for skill in skills:
    df[skill] = df['Job Description'].apply(lambda row: 1 if skill in row.lower() else 0)

```
